```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
from yellowbrick.cluster import KElbowVisualizer, SilhouetteVisualizer
```

```python
filepath = '/Users/efkanturedi/Corteze/Python_output_folders/Ecommerce/final.csv'

master_df_clean = pd.read_csv(filepath)
```

```python
master_df_clean
```

```python
master_df_clean = master_df_clean.set_index('customer_unique_id')
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans()
visualizer = KElbowVisualizer(model, k=(4,15))

visualizer.fit(master_df_clean)    # Fit the data to the visualizer
visualizer.poof() 
```

```python
model = KMeans(8)
visualizer = SilhouetteVisualizer(model)

visualizer.fit(master_df_clean)    # Fit the data to the visualizer
visualizer.poof()    # Draw/show/poof the data
```

```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN()
visualizer = KElbowVisualizer(model, k=(4,15))

visualizer.fit(master_df_clean)    # Fit the data to the visualizer
visualizer.poof()
```
