<h2> K-means Analysis

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer, SilhouetteVisualizer
```

```python
model = KMeans()
visualizer = KElbowVisualizer(model, k=(60,80),metric='silhouette')

visualizer.fit(df_clean)    # Fit the data to the visualizer
visualizer.poof()    # Draw/show/poof the data
```

```python
visualizer = KElbowVisualizer(model, k=(12,20))
visualizer.fit(df_clean)    # Fit the data to the visualizer
visualizer.poof()  
```

```python
model = KMeans(16)
visualizer = SilhouetteVisualizer(model)

visualizer.fit(df_clean)    # Fit the data to the visualizer
visualizer.poof() 
```

<h2> K-Means / T-SNE Visualisation

```python
from sklearn.manifold import TSNE
```

```python
optimal_parameter = 16
k_means = KMeans(optimal_parameter)
k_means.fit(df_clean.copy())
clusters_k_means = k_means.labels_
```

```python
tsne = TSNE()
T = tsne.fit_transform(df_clean.copy())
```

```python
plt.figure(figsize=(8,8))
plt.title('Représentation du jeu de données via T-SNE')
sns.scatterplot(T[:,0], T[:,1],hue=clusters_k_means,legend='full',palette="hls")
```

Regarder les centroids. Regarder les dimensions principales (argmax) via les coordonnées des centroides.  

<h2> DBSCAN: finding the optimal parameters

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
```

```python
dbs = DBSCAN(eps = 0.2, 
             min_samples=3,
               n_jobs=-1).fit(df_clean.copy())
dbs.labels_
```

```python
silhouette_score(df_clean,dbs.labels_)
```

```python
tsne = TSNE()
T = tsne.fit_transform(df_clean.copy())
```

```python
plt.figure(figsize=(8,8))
plt.title('Représentation du jeu de données via T-SNE')
sns.scatterplot(T[:,0], T[:,1],hue=dbs.labels_,legend='full',palette="hls")
```

```python
cat_df_clean = df_clean.assign(Cluster=dbs.labels_)
```

```python
cat_df_clean['Cluster'].value_counts().sort_values()
```

```python
df = df.assign(Cluster_k_means=k_means.labels_,Cluster_DBS=dbs.labels_)
```
