```python
df = master_df[[
    'customer_unique_id',
    'order_id',
    'order_status',
    'order_purchase_date',
    'product_id',
    'order_item_id',
    'price',
    'product_category_name'
    ]].copy()
```

```python
df['order_purchase_date'] = pd.to_datetime(df['order_purchase_date'], format='%Y%m%d') 
```

```python
# Setting reference day
df['today'] = df['order_purchase_date'].max()
# Date deltas
df['recency'] = df['today'] - df['order_purchase_date']
```

```python
df = df[df['order_status']=='delivered']
df.reset_index(drop=True,inplace=True)
```

```python
df['price'] = pd.to_numeric(df['price'])
```

```python
df['price'].dtype
```

```python
df['total_price'] = df['order_item_id'] * df['price']
```

```python
# Grouping by customer and order (same order_id = same day); This will be the data frame which contains all valid costomers and how they are segmented
df_segmented = df.groupby([
    'customer_unique_id',
    'order_id'
    ], as_index=False).agg({
        'order_purchase_date': 'first',
        'order_item_id': 'max',
        'total_price': 'sum',
        'today': 'first',
        'recency': 'first'
        })
```

```python
df_segmented
```

```python
df_segmented['average_price'] = df_segmented['total_price'] / df_segmented['order_item_id']
```

```python
df_segmented['recency'] = df_segmented['recency'].astype(str)
df_segmented['recency'] = df_segmented['recency'].str.replace(r'\s+days.*', '', regex=True)
df_segmented['recency'] = df_segmented['recency'].astype(int)
```
