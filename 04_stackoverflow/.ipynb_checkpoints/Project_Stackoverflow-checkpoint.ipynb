<h4> Tokenisation de la colonne Title

```python
from nltk.corpus import stopwords
from nltk import word_tokenize
from nltk.tokenize import sent_tokenize
```

```python
stop_words = set(stopwords.words('english'))
```

```python
regularizer = lambda x : re.sub("[^a-zA-Z]",           # The pattern to search for
                      " ",                   # The pattern to replace it with
                      x)
```

```python
db['Title'] = list(map(regularizer, db['Title']))
```

```python
db
```

```python
lower_case = lambda x : x.lower()
db['Title'] = list(map(lower_case,db['Title']))
```

```python
db['Title']
```

```python
db['Title'] = list(map(word_tokenize,db['Title']))
```

```python
db
```

```python
db['Title'] = [stop_words_check(word_list,stop_words) for word_list in db['Title']]
```

```python
db
```

<h4> Lemmatization des tokens


```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
def list_lemmatizer(list_to_lem):
    return [lemmatizer.lemmatize(token, 'v') for token in list_to_lem]
```

```python
db['Title'] = list(map(list_lemmatizer,db['Title']))
```

```python
db
```

```python
db['Title'] = list(map(" ".join,db['Title']))
```

```python
db
```

<h2> Approche Supervisée

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.preprocessing import OneHotEncoder
```

```python
X = np.array(db['Title'])
y = np.array(db['Tags'])
```

```python
ohe = OneHotEncoder()
```

```python
y = y.reshape(-1,1)
y_encoded = ohe.fit_transform(y)
```

```python
vectorizer_X1 = TfidfVectorizer(analyzer = 'word',
                                       min_df=0.0,
                                       max_df = 1.0,
                                       strip_accents = None,
                                       encoding = 'utf-8', 
                                       preprocessor=None,
                                       token_pattern=r"(?u)\S\S+",
                                       max_features=1000)
```

```python
X1_tfidf = vectorizer_X1.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X1_tfidf, y_encoded, test_size = 0.2, random_state = 0) # Do 80/20 split
```

Nous sommes maintenant prêt à faire tourner les modèles

```python
from sklearn.dummy import DummyClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import jaccard_score
from sklearn.metrics import accuracy_score
from sklearn.linear_model import SGDClassifier, LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
```

<h4> One vs Rest

```python
dummy = DummyClassifier()
sgd = SGDClassifier()
lr = LogisticRegression()
svc = LinearSVC()

for classifier in [dummy, sgd, lr, svc]:
    clf = OneVsRestClassifier(classifier)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    #print(jaccard_score(y_test, y_pred, average= 'weighted'))
    print(accuracy_score(y_test, y_pred))
```

<h4> Random Forest

```python

#rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)

y_pred = rfc.predict(X_test)

print(accuracy_score(y_test, y_pred, average= 'weighted'))
```

<h4> Multilayered Perceptron

```python
mlpc = MLPClassifier()
mlpc.fit(X_train, y_train)

y_pred = mlpc.predict(X_test)

print(jaccard_score(y_test, y_pred, average= 'weighted'))
print(accuracy_score(y_test, y_pred))
```
